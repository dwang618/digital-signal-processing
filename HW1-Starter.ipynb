## Question 3:

Determine if each of the following systems is (1) stable, (2) causal, (3) linear, and (4) time invariant. Note: a system may have multiple of these properties. **Briefly explain why each property does or does not hold.**

## Answer:

## (a)

    **(1) No, it is not stable.**

The BIBO stability of this time system is dependent on whether the input is bounded relative to its output. For any discrete signal $x[n]$ for any integer 'n', the resulting system $x[n^3]$ grows at an exponential rate and is not bounded by its cubic relationship.

    **(2) No, it is not causal.**

The system is not causal, such that $T\{x[n]\} = x[n^3]$ can be represented as $T\{x[2]\} = x[2^3]$ where the system is dependent on proceeding values:

$n=2 < n^3 = 2^3 = 8$.

    **(3) Yes, it is linear.**
    
The system is linear because it fulfills both the additivity property and the scaling property. 

$T\{ax[n]\} = aT\{x[n]\} = ax[n^3] $ 

$T\{x_1[n] + x_2[n]\} = T\{x_1[n]\} + T\{x_2[n]\} = x_1[n^3] + x_2[n^3]$

    **(4) Yes, it is time invariant.**
The system is time invariant because it satisfies the property of being shift invariant for any integer value of displacment 'N', such that: 

$y[n] = T\{x[n]\} \rightarrow y[n-N] = T\{x[n-N]\}$

$x[n^3] = T\{x[n]\} \rightarrow x[(n-N)^3] = T\{x[n-N]\}$

    **(2) Yes, it is causal.**

The system is causal, such that $T\{x[n]\} = e^{i\pi n} x[n]$ is dependent on only present values corresponding with $x[n]$. While the system does contain a complex component, it has no effect on the relationship between 'n' and its output through the system.

```python
import numpy as np  
import matplotlib.pyplot as plt
import wave
from IPython.display import Audio

def FFCF(inputSignal, delayVal, gainVal):
    inputArray = np.asarray(inputSignal)
    outputArray = np.zeros_like(inputArray)
    delayVal = (int) (delayVal)
    gainVal = (float) (gainVal)
    
    size = len(inputArray)
    for n in range(size):
        shiftIndex = n-delayVal
        if shiftIndex >= 0:
            outputArray[n] = inputArray[shiftIndex] * gainVal + inputArray[n]
        else:
            outputArray[n] = inputArray[n]
            
    return outputArray

def FBCF(inputSignal, delayVal, gainVal):
    inputArray = np.asarray(inputSignal)
    outputArray = np.zeros_like(inputArray)
    delayVal = (int) (delayVal)
    gainVal = (float) (gainVal)
    
    size = len(inputArray)
    for n in range(size):
        outputArray[n] = inputArray[n]
        if n - delayVal >= 0:
            outputArray[n] = inputArray[n] + outputArray[n-delayVal] * gainVal
            
    return outputArray
```
