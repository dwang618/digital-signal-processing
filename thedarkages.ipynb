* **(c)** Implement both the FFCF and FBCF systems as Python functions. Each of your functions should take three parameters: (1) the input signal $x[n]$ (i.e., a `numpy` array of any length), (2) the delay $N > 0$, and (3) the {\em gain}, $g \in \mathbb{R}$. * **(d)** Run both of your systems on the signal `woodchuck.wav` with a delay of 100ms and a gain of 0.75. Note: you will have to convert the delay from milliseconds to number of samples, $N$. Play the resulting output audio signal and describe what you hear. What are the differences between the resulting sounds from the FFCF and FBCF outputs? (Just looking for your qualitative observations.)

The FFCF system contains a noticeable echo where the input is repeated at a slight delay. While the difference is significant, the system does not entirely distort the original input and all the words can still be made out by ear. The FBCF system contains a similar effect that is in some ways an amplified version of the FCFF system's behavior. There seems to be an increase of echoing signals that cause the audio to wave longer and more frequently.


```python
import numpy as np  
import matplotlib.pyplot as plt
import wave
from IPython.display import Audio

def FFCF(inputSignal, delayVal, gainVal):
    inputArray = np.asarray(inputSignal)
    outputArray = np.zeros_like(inputArray)
    delayVal = (int) (delayVal)
    gainVal = (float) (gainVal)
    
    size = len(inputArray)
    for n in range(size):
        shiftIndex = n-delayVal
        if shiftIndex >= 0:
            outputArray[n] = inputArray[shiftIndex] * gainVal + inputArray[n]
        else:
            outputArray[n] = inputArray[n]
            
    return outputArray

def FBCF(inputSignal, delayVal, gainVal):
    inputArray = np.asarray(inputSignal)
    outputArray = np.zeros_like(inputArray)
    delayVal = (int) (delayVal)
    gainVal = (float) (gainVal)
    
    size = len(inputArray)
    for n in range(size):
        outputArray[n] = inputArray[n]
        if n - delayVal >= 0:
            outputArray[n] = inputArray[n] + outputArray[n-delayVal] * gainVal
            
    return outputArray
```

```python
def wavToFloatArr(filepath):
    # Open the WAV file
    wave_audio = wave.open(filepath, 'rb')
    
    # Get audio parameters
    framerate = wave_audio.getframerate()
    num_frames = wave_audio.getnframes()

    # Read and convert audio data to a float array
    audio_data = np.frombuffer(wave_audio.readframes(num_frames), dtype=np.int16)
    audio_data = audio_data.astype(np.float32) / 32767.0  # Normalize to [-1, 1]

    # Close the WAV file
    wave_audio.close()
    return audio_data, framerate

wave_audio = wave.open(r"C:\Users\Dwang\Documents\school\dsp\woodchuck.wav")
passingWaveAudio = r"C:\Users\Dwang\Documents\school\dsp\woodchuck.wav"
woodchuck_signal, framerate = wavToFloatArr(passingWaveAudio)

delayValue = framerate // 10
output_ffcf = FFCF(woodchuck_signal, delayValue, gainVal=0.75)
output_fbcf = FBCF(woodchuck_signal, delayValue, gainVal=0.75)

Audio(output_ffcf, rate=framerate)
```

* **(j)** Create another unit impulse, now with length equal to the `woodchuck.wav` signal. Compute the corresponding impulse response, $h[n]$, for the FBCF system with delay = 100ms, gain = 0.75. Now use your convolve function to convolve `woodchuck.wav` with $h[n]$. Play the audio---does it match what you heard in part (d)?

```python
def convolve_signals(x, h):
    # Convert input signals to NumPy arrays
    x_input = np.asarray(x)
    h_input = np.asarray(h)
    
    # Result that is the same length as input1 for correct array size
    result_conv_signal = np.zeros(x_input.size + h_input.size - 1)

    # Zero pad the input signal x before convolution
    input_x_padded = np.concatenate((np.zeros(len(h_input) - 1), x_input, np.zeros(len(h_input) - 1)))

    # Reverse input signal h, effectively flip across x axis
    h_input = np.flip(h)

    # Convolution process: Iterate through positions, calculate convolution result by summing element-wise products 
    #of input segment and flipped filter
    for i in range(result_conv_signal.size):
        start_idx = i
        end_idx = i + len(h_input)
        # Calculate convolution for the current position
        result_conv_signal[i] = np.sum(input_x_padded[start_idx:end_idx] * h_input)

    return result_conv_signal
```

```python
delayValue = framerate // 10
unit_impulse = np.zeros_like(woodchuck_signal)
unit_impulse[0] = 1

#Parameters
delay_ms = 100
delay_samples = int(delayValue * framerate / 1000)  # Convert delay to samples
gain = 0.75

#FBCF system with delay 
impulse_response = FBCF(unit_impulse, delay_samples, gain)
convolved_audio = convolve_signals(woodchuck_signal, impulse_response)

```
